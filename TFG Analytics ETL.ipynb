```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from datetime import datetime
from tqdm import tqdm
import time

client_id = '9645b67041ed4d1ab11ca804cf9172ff'
client_secret = '1f01a881f7724afd83a4f6dc32089098'

client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

df = pd.read_csv('/Users/cres/european_countries_top_50_audio_characteristics_updated_final(2).csv') 

def safe_spotify_request(func, *args, **kwargs):
    for _ in range(5):  # Número de reintentos
        try:
            return func(*args, **kwargs)
        except spotipy.exceptions.SpotifyException as e:
            if e.http_status == 429:  # Límite de tasa alcanzado
                time.sleep(int(e.headers.get('Retry-After', 1)))  # Espera el tiempo recomendado
            else:
                break  
        except Exception as e:
            print(f"Error inesperado: {e}")
            break
    return None

def get_artist_followers(artist_id):
    if artist_id:
        return safe_spotify_request(sp.artist, artist_id).get('followers', {}).get('total')

from datetime import datetime

def parse_date(date_str):
    for fmt in ('%Y-%m-%d', '%Y-%m', '%Y'):
        try:
            return datetime.strptime(date_str, fmt)
        except ValueError:
            pass
    raise ValueError('no valid date format found')

def get_track_info(track_id):
    if track_id:
        track_info = safe_spotify_request(sp.track, track_id)
        if track_info:
            artist_id = track_info['artists'][0]['id']
            genres = safe_spotify_request(sp.artist, artist_id).get('genres', [])
            release_date_str = track_info['album']['release_date']
            release_date = parse_date(release_date_str)
            return genres, release_date
    return [], None

df['followers'] = None
df['genres'] = None
df['release_date'] = None
df['time_past_since_release'] = None

for index, row in tqdm(df.iterrows(), total=df.shape[0], desc="Procesando"):
    df.at[index, 'followers'] = get_artist_followers(row['artist_id'])
    genres, release_date = get_track_info(row['track_id'])
    df.at[index, 'genres'] = genres
    df.at[index, 'release_date'] = release_date
    if release_date:
        df.at[index, 'time_past_since_release'] = (datetime(2023, 12, 1) - release_date).days

df.drop(columns=['date'], inplace=True)
df.rename(columns={'release_date': 'date'}, inplace=True)

df.to_csv('/Users/cres/european_countries_top_50_audio_characteristics_updated_final(3).csv', index=False)
```

```python
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

client_id = '9645b67041ed4d1ab11ca804cf9172ff'
client_secret = '1f01a881f7724afd83a4f6dc32089098'
client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

df = pd.read_csv('/Users/cres/eurovision_2023_participants.csv')

# Función para buscar el track_id y características de audio en Spotify

def get_track_data(track_name, artist_name):
    query = f"track:{track_name} artist:{artist_name}"
    results = sp.search(q=query, type='track', limit=1)
    items = results['tracks']['items']
    if not items:
        return None, None, None  # Se devuelven tres valores None si no hay resultados
    
    track_id = items[0]['id']
    popularity = items[0]['popularity']
    audio_features = sp.audio_features(tracks=[track_id])[0]
    
    return track_id, popularity, audio_features

# Aplicar la función a cada fila y crear las columnas 'track_id', 'popularity' y audio features
for index, row in df.iterrows():
    track_id, popularity, audio_features = get_track_data(row['track_name'], row['artist_name'])
    df.at[index, 'track_id'] = track_id
    df.at[index, 'popularity'] = popularity
    if audio_features:
        for feature in audio_features:
            df.at[index, feature] = audio_features[feature]

df.to_csv('/Users/cres/eurovision_2023_participants_with_audio_features.csv', index=False)
```
